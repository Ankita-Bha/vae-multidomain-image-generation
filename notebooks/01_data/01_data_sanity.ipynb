```python
from pathlib import Path

# Robustly locate project root
current = Path().resolve()
while not (current / "src").exists():
    current = current.parent

print("Project root:", current)

# Check grayscale checkpoints
ckpt_dir = current / "checkpoints" / "grayscale"
checkpoints = list(ckpt_dir.glob("*.pt"))

print("Found checkpoints:")
for ckpt in checkpoints:
    print(" -", ckpt)

print("\nTotal checkpoints:", len(checkpoints))
```
